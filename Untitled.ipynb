```python
import time
from datetime import datetime, timezone
from pathlib import Path

import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt

%matplotlib inline
pd.set_option("display.max_columns", None)
```

```python
SYMBOL = "BTCUSDT"
INTERVAL = "5m"          # 1m, 3m, 5m, 15m, 30m, 1h, 4h, 1d, ...
START_DATE = "2023-07-23"  # UTC, inclusive -- 3 years back from today
END_DATE = "2026-07-23"    # UTC, inclusive

MAX_FFILL_CANDLES = 3    # max consecutive missing candles to forward-fill during cleaning

RAW_DATA_PATH = Path(f"data/raw/{SYMBOL}_{INTERVAL}_raw.csv")
CLEAN_DATA_PATH = Path(f"data/processed/{SYMBOL}_{INTERVAL}_clean.parquet")
EDA_OUTPUT_DIR = Path("eda_output")

RAW_DATA_PATH.parent.mkdir(parents=True, exist_ok=True)
CLEAN_DATA_PATH.parent.mkdir(parents=True, exist_ok=True)
EDA_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
BASE_URL = "https://api.binance.com"
KLINES_ENDPOINT = "/api/v3/klines"
MAX_CANDLES_PER_REQUEST = 1000
REQUEST_SLEEP_SECONDS = 0.3

KLINE_COLUMNS = [
    "open_time", "open", "high", "low", "close", "volume",
    "close_time", "quote_asset_volume", "num_trades",
    "taker_buy_base_volume", "taker_buy_quote_volume", "ignore",
]

INTERVAL_MS = {
    "1m": 60_000, "3m": 3 * 60_000, "5m": 5 * 60_000, "15m": 15 * 60_000,
    "30m": 30 * 60_000, "1h": 60 * 60_000, "4h": 4 * 60 * 60_000, "1d": 24 * 60 * 60_000,
}


def to_ms(date_str: str) -> int:
    dt = datetime.strptime(date_str, "%Y-%m-%d").replace(tzinfo=timezone.utc)
    return int(dt.timestamp() * 1000)


def fetch_klines(symbol, interval, start_ms, end_ms, session=None, max_retries=5):
    """Fetch all klines between start_ms and end_ms (inclusive), paginating as needed."""
    if interval not in INTERVAL_MS:
        raise ValueError(f"Unsupported interval '{interval}'. Known: {list(INTERVAL_MS)}")

    session = session or requests.Session()
    step_ms = INTERVAL_MS[interval]
    all_rows = []
    cursor = start_ms

    while cursor <= end_ms:
        params = {
            "symbol": symbol, "interval": interval,
            "startTime": cursor, "endTime": end_ms,
            "limit": MAX_CANDLES_PER_REQUEST,
        }

        rows = None
        for attempt in range(max_retries):
            try:
                resp = session.get(BASE_URL + KLINES_ENDPOINT, params=params, timeout=10)
                if resp.status_code == 429:
                    wait = 2 ** attempt
                    print(f"  Rate limited (429). Backing off {wait}s...")
                    time.sleep(wait)
                    continue
                resp.raise_for_status()
                rows = resp.json()
                break
            except requests.RequestException as exc:
                wait = 2 ** attempt
                print(f"  Request error ({exc}). Retrying in {wait}s...")
                time.sleep(wait)

        if rows is None:
            raise RuntimeError(f"Failed to fetch klines starting at {cursor} after {max_retries} retries")
        if not rows:
            break

        all_rows.extend(rows)
        last_open_time = rows[-1][0]
        next_cursor = last_open_time + step_ms
        if next_cursor <= cursor:
            break
        cursor = next_cursor

        print(f"  Fetched {len(rows):4d} candles up to "
              f"{datetime.fromtimestamp(last_open_time / 1000, tz=timezone.utc)} "
              f"(total so far: {len(all_rows)})")
        time.sleep(REQUEST_SLEEP_SECONDS)

    return all_rows


def klines_to_dataframe(rows):
    """Convert raw kline rows into a typed, tidy DataFrame."""
    df = pd.DataFrame(rows, columns=KLINE_COLUMNS)

    numeric_cols = ["open", "high", "low", "close", "volume",
                     "quote_asset_volume", "num_trades",
                     "taker_buy_base_volume", "taker_buy_quote_volume"]
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    df["open_time"] = pd.to_datetime(df["open_time"], unit="ms", utc=True)
    df["close_time"] = pd.to_datetime(df["close_time"], unit="ms", utc=True)

    df = df.drop(columns=["ignore"])
    df = df.sort_values("open_time").reset_index(drop=True)
    df = df.drop_duplicates(subset="open_time", keep="first").reset_index(drop=True)
    return df
```

```python
start_ms = to_ms(START_DATE)
end_ms = to_ms(END_DATE)

print(f"Fetching {SYMBOL} {INTERVAL} candles from {START_DATE} to {END_DATE}...")
raw_rows = fetch_klines(SYMBOL, INTERVAL, start_ms, end_ms)
print(f"Fetched {len(raw_rows)} raw candles total.")

df_raw = klines_to_dataframe(raw_rows)
df_raw.to_csv(RAW_DATA_PATH, index=False)
print(f"Saved {len(df_raw)} candles to {RAW_DATA_PATH}")
df_raw.head()
```
